## üìä Forme de la distribution

```python
# Nombre de graphiques par ligne
NB_COLS = 3

plt.figure(figsize=(15, 15))

# S√©lection des colonnes explicatives uniquement
# On exclut :
# - 'id' : identifiant technique (aucune valeur pr√©dictive)
# - 'diagnosis' : variable cible (utilis√©e pour la couleur des graphiques)
selected_cols = df.drop(["id", "diagnosis"], axis=1).columns

# Enumeration des colonnes pour positionner chaque graphique
for i, col in enumerate(selected_cols, 1):

    # Cr√©ation d'un sous-graphique dans une grille
    # Nombre de lignes calcul√© automatiquement
    plt.subplot(round(len(selected_cols) / NB_COLS), NB_COLS, i)

    # Histogramme de la variable
    # 'hue=diagnosis' permet de comparer les distributions
    # entre tumeurs b√©nignes (B) et malignes (M)
    sns.histplot(
        data=df,
        x=col,
        hue="diagnosis"
    )
    plt.xlabel(col)
    
plt.tight_layout()
plt.show()
```

## üîç √âtudes et gestion des outliers 

```python
# S√©lection des variables les plus corr√©l√©es avec la cible 'diagnosis'
selected_features = [
    "area_worst",
    "radius_worst",
    "perimeter_worst",
    "concavity_mean",
    "concave points_mean"
]

# DataFrame qui va contenir TOUS les outliers d√©tect√©s
outliers_df = pd.DataFrame()

print("üö® D√©tection des outliers (m√©thode IQR)")
print("=" * 40)

# Boucle sur chaque variable s√©lectionn√©e
for col in selected_features:
    
    Q1 = df[col].quantile(0.25)  
    Q3 = df[col].quantile(0.75)  
    IQR = Q3 - Q1                

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_col = df[(df[col] < lower_bound) | (df[col] > upper_bound)].copy()

    # Ajout d'une colonne pour savoir quelle variable a d√©clench√© l'outlier
    outliers_col["outlier_variable"] = col

    # Ajout des outliers d√©tect√©s pour cette variable au DataFrame global
    outliers_df = pd.concat([outliers_df, outliers_col], axis=0)


# Suppression des doublons (un m√™me individu peut √™tre outlier sur plusieurs variables)
outliers_df = outliers_df.drop_duplicates()


print(f"Nombre total d'individus pr√©sentant des observations atypiques : {outliers_df.shape[0]}")

outliers_df
```

## üìå Interpr√©tation m√©tier des outliers

Les individus identifi√©s comme outliers pr√©sentent :

- soit des **dimensions tumorales extr√™mes**
- soit des **anomalies morphologiques marqu√©es**

Ces observations sont coh√©rentes avec des **profils potentiellement malins**.

Ils seront :
- inclus dans l‚Äôentra√Ænement du mod√®le
